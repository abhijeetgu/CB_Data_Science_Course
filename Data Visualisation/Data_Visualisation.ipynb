## Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
plt.plot(x,y1,color = 'red',label='Apple',marker='o')
#plt.show()
plt.plot(x,y2,color = 'green',label="Kiwi",linestyle='dashed',marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.legend()
plt.show()
```

```python
prices  = np.array([1,2,3,4])**3
print(prices)
```

```python
plt.plot(prices)
plt.show()
```

## Scatter Plots

```python
# Adjust the size of any plot
plt.figure(figsize=(6,6))
plt.scatter(x,y1,color = 'red',label='Apple',marker='^')
#plt.show()
plt.scatter(x,y2,color = 'green',label="Kiwi",linestyle='dashed',marker='*')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot-Prices of Fruits over Time")
plt.legend()
plt.show()
```

## Bar Graph

```python
plt.style.use("classic")
x_cordinates = np.array([0,1,2])*2
plt.bar(x_cordinates-0.25,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"]) #Current year
plt.bar(x_cordinates+0.25,[20,10,12],width=0.5,label = "Next Year") #Next Year
plt.title("Metal Prices comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_cordinates = np.array([0,1,2])*2
plt.bar(x_cordinates-0.25,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platinum","Silver"],color="yellow") #Current year
plt.bar(x_cordinates+0.25,[20,10,12],width=0.5,label = "Next Year",color="orange") #Next Year
plt.title("Metal Prices comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.legend()
plt.show()
```

## Pie Chart

```python
plt.style.use("seaborn")
subjects = "Maths","Chemistry","Physics","English"
weightage = [20,10,15,5]
plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0),autopct='%1.1f%%',shadow=True)
plt.title("Subjects")
plt.show()
```

## Histogram

```python
#Xsn (standared normal distribution)
Xsn = np.random.randn(100)
sigma = 8
u = 70
X = np.round(Xsn*sigma + u) #For physics marks
X2 = np.round(Xsn*7 + 40)
print(X,X2)
```

```python
plt.hist(X,alpha=0.8,label ="Physics")
plt.hist(X2,alpha=0.9,label = "Maths")
plt.xlabel("Marks Range")
plt.ylabel("Prob/FreqCount of students")
plt.title("Histogram")
plt.legend()
plt.show()
```
